PDF to txt

```python
import os
# from PyPDF2 import PdfReader
import pdfplumber

input_pdf = "/content/drive/MyDrive/Dash-7b/Dash-7b/qa/mml-book.pdf"
output_txt = "/content/drive/MyDrive/Dash-7b/Dash-7b/qa/mml-book.txt"

with pdfplumber.open(input_pdf) as pdf:
    num_pages = len(pdf.pages)

    with open(output_txt, "w", encoding="utf-8") as txt_file:
        for page_number in range(num_pages):
            page = pdf.pages[page_number]
            text = page.extract_text()
            txt_file.write(text + os.linesep)
```

transcript to txt

```python
import csv

def append_transcripts_to_txt(csv_file, txt_file):
    with open(csv_file, 'r', encoding='utf-8') as f_csv, open(txt_file, 'a', encoding='utf-8') as f_txt:
        reader = csv.DictReader(f_csv)
        for row in reader:
            transcript = row['transcript']
            f_txt.write(transcript + '\n')

csv_file_path = '/content/drive/MyDrive/Dash-7b/Dash-7b/sum/sumdata.csv'
txt_file_path = '/content/drive/MyDrive/Dash-7b/Dash-7b/qa/transcript.txt'

append_transcripts_to_txt(csv_file_path, txt_file_path)
```
